```python
#1.create a losr of 7 flowers and perform label encoding on them.
from sklearn.preprocessing import LabelEncoder
flowers =["Rose","Sunflower","Lily","Lotus","Orchid","dahlia","Marigold"]
encoder = LabelEncoder()
flowers_encoded = encoder.fit_transform(flowers)
print(flowers_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
#2.Create a dataframe with 5 column (categorical+numerical)and perform one hor encoding.
import pandas as pd
df = pd.DataFrame({"Fruits":["Apple","Cherry","Banana","Grapes","Cherry","Apple"],
                   "Season":["Winter","Manson","Winter","Summer","Winter","Manson"],
                   "Color":["Red","Yellow","Green","Red","Yellow","Green"],
                   "Weight_g":[150,120,180,160,150,130],
                   "Price_rs":[50,30,60,70,30,60]
})
df = pd.DataFrame(df)
print(df)

encoded_df = pd.get_dummies(df,columns = ["Fruits","Season","Color","Weight_g","Price_rs"],dtype=int)
print(encoded_df)
```

```python
#3.create a list of ordinal data with 6 different values and perform ordinal encoding on them.
from sklearn.preprocessing import OrdinalEncoder
size = [["Small"],["Medium"],["Large"],["Extra Large"],["Double XL"],["Triple XL"],["Small"],["Medium"],["Large"],["Extra Large"],["Double XL"],["Triple XL"]]
encoder = OrdinalEncoder(categories = [["Small","Medium","Large","Extra Large","Double XL","Triple XL"]])
encoded_size = encoder.fit_transform(size)
print(encoded_size)
```
