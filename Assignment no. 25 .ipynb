```python
# Assignment no.25
#1. Extract only the alphabetic characters from a string containing letters, digits, and symbols.
a = input("Enter a string: ")
result = ""
for ch in a:
    if ch.isalpha():  
        result += ch
print("Alphabetic characters:", result)
```

```python
#2. Normalize a string so multiple spaces between words become a single space.
a = input("Enter a string: ")
words = a.split()  
result = " ".join(words)
print(result)
```

```python
#3. Check if two strings are anagrams while ignoring case and spaces.
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
s1 = s1.replace(" ", "").lower()
s2 = s2.replace(" ", "").lower()
if sorted(s1) == sorted(s2):
    print("Strings are anagrams")
else:
    print("Strings are not anagrams")
```

```python
#4. Reverse a string without using slicing.
string = input()
rev_string= ""
for ch in string:
    rev_string = ch + rev_string
print(rev_string)
```

```python
#5. Count the number of sentences in a paragraph ending with '.' , '?' , or '!' .
text = input("Enter a paragraph: ")
count = 0
end_marks = ['.', '?', '!']
for ch in text:
    if ch in end_marks:
        count += 1
print("Number of sentences:", count)
```

```python
#7. Compress a string like 'aaabbccccdde' into 'a3b2c4d2e1'.
s = input("Enter a string: ")
result = ""
count = 1
for i in range(len(s)):
    if i < len(s)-1 and s[i] == s[i+1]:
        count += 1
    else:
        result += s[i] + str(count)
        count = 1
print(result)
```

```python
#8. Check if a string contains a repeating substring pattern.
s = input("Enter a string: ")
if s in (s + s)[1:-1]:
    print("String has a repeating substring pattern")
else:
    print("String does NOT have a repeating substring pattern")
```

```python
#9. Remove characters that appear more than once in a string.
s = input("Enter a string: ")
result = ""
for ch in s:
    if s.count(ch) == 1:  
        result += ch
print("Result:", result)
```

```python
#10Split a camelCase string into separate words.
s = input("Enter a camelCase string: ")
result = ""
for ch in s:
    if ch.isupper():
        result += " " + ch
    else:
        result += ch
print(result.strip())
```

```python
#11. Rotate a string to the right by k positions using slicing only once.
s = input("Enter a string: ")
k = int(input("Enter k: "))
k = k % len(s)    
rotated = s[-k:] + s[:-k]
print("Rotated string:", rotated)
```

```python
#16. Convert a string into alternating case starting with uppercase.
s = input("Enter a string:")
count = ""
i = 0
for ch in s:
    if i % 2 == 0:
        count += ch.upper()
    else:
        count += ch.lower()
    i += 1
print(count)
```

```python
#17. Count words in a sentence that start and end with the same letter.
s = input("Enter a sentence: ")
count = 0
for w in s.split():
    if w[0].lower() == w[-1].lower():
        count += 1
print(count)
```

```python
#18. Swap every pair of characters in a string.
s = input("Enter a string: ")
result = ""
for i in range(0, len(s), 2):
    pair = s[i:i+2]       
    result += pair[::-1]  
print(result)
```

```python
#19. Clean comma-separated emails: trim, lowercase, unique, sorted.
s = input("Enter emails: ")
emails = s.split(",")
clean = sorted(set(e.strip().lower() for e in emails))
print(",".join(clean))
```

```python
#24. Rearrange a string so lowercase letters appear first, then uppercase.
s = input("enter a string:")
lower = ""
upper = ""
for c in s:
    if c.islower():
        lower += c
    else:
        upper += c
print(lower + upper)
```

```python
#25. Generate all unique subsets of characters in lexicographical order.
s = input("Enter a string:")
s = "".join(sorted(set(s)))
r = [""]
for ch in s:
    r += [x + ch for x in r]

print(sorted(r))
```
