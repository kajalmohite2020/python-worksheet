```python
#11. Perform multiple aggregations on course to calculate average quiz score, average assignment score, and average project score.
df.groupby("course")[["avg_quiz_score","assignment_score","project_score"]].agg("mean")
```

The code performs multiple aggregations on the “course” column to compute three average scores are show in the four column like Average quiz score – the mean score students got in quizzes for each course. Average assignment score – the mean score students achieved on assignments per course.Average project score – the mean score students earned on projects for each course.

```python
#12. Perform named aggregations on course to calculate average_attendance, max_project_score, and mean_total_score.
df["total_score"] = df["avg_quiz_score"] + df["project_score"]
result = df.groupby("course").agg(
    average_attendance=("attendance_pct", "mean"),
    max_project_score=("project_score", "max"),
    mean_total_score=("total_score", "mean"))
df.head()
```

The total_score column is the sum of avg_quiz_score and project_score, representing a student's overall performance in the course.
