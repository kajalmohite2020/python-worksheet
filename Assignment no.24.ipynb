```python
#Assignment no.22
#1. Write a Python program that takes a password as input and checks if it meets the following requirements:
#- The password should be at least 8 characters long.
#- The password should contain at least one uppercase letter.
#- The password should contain at least one lowercase letter.
#- The password should contain at least one digit.
#If the password meets all the requirements, the program should print "Password is valid." Otherwise, it should print "Password is invalid."
n=input("enter a password:")
upper=False
Lower=False
Digit=False
count=0
for i in n:
    count +=1
    if i.isupper():
        upper=True
    if i.islower():
        Lower=True
    if i.isdigit():
        Digit=True
if count>=8 and upper and Lower and Digit:
    print("Password is valid")
else:
    print("Password is invalid")
```

```python
#2. Write a Python program that takes two strings as input and checks if they're anagrams (i.e., the strings contain the same characters, but in a different order). If the strings are anagrams, the program should print "The strings are anagrams." Otherwise, it should print "The strings are not anagrams."
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
if sorted(s1) == sorted(s2):
    print("The strings are anagrams.")
else:
    print("The strings are not anagrams.")
```

```python
# 3. Given a 3-letter string, classify its pattern:
# AAA if all letters same
# ABA if outer same
# ABC if all different

a = input("Enter a 3-letter string: ")
if len(a) == 3:
    if a[0] == a[1] == a[2]:
        print("AAA(all letters same)")
    elif a[0] == a[2]:
        print("ABA(if outer same)")
    else:
        print("ABC(if all different)")
```
